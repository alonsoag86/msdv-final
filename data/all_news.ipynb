```python
media_month = d.groupby(['media_outlet', 'month']).size().rename('media_month')
print(media_month)
```

```python
d2 = d.join(media_month, on=['media_outlet','month'])
```

```python
media_year = d2.groupby(['media_outlet']).size().rename('media_year')
```

```python
print(media_year)
```

```python
d3 = d2.join(media_year, on=['media_outlet'])
```

```python
d3.head()
```

```python
tendency_count_month = d3.groupby(['month','tendency']).size().rename('tendency_month')
print(tendency_count_month)
```

```python
d4 = d3.join(tendency_count_month, on=['month','tendency'])
```

```python
d4.head()
```

```python
tendency_count_year = d4.groupby(['tendency']).size().rename('tendency_year')
print(tendency_count_year)
```

```python
d5 = d4.join(tendency_count_year, on=['tendency'])
```

```python
theme_month = d5.groupby(['theme','month']).size().rename('theme_month')
print(theme_month)
```

```python
d6 = d5.join(theme_month, on=['theme','month'])
```

```python
theme_year = d6.groupby(['theme']).size().rename('theme_year')
print(theme_year)
```

```python
d7 = d6.join(theme_year, on=['theme'])
```

```python
month_count = d7.groupby(['month']).size().rename('month_count')
print(month_count)
```

```python
d8 = d7.join(month_count, on=['month'])
```

```python
d8['total_count'] = 7815
```

```python
d8['tendency_year_percentage'] = d8['tendency_year']/7815
d8['theme_year_percentage'] = d8['theme_year']/7815
d8['tendency_month_percentage'] = d8['tendency_month']/d8['month_count']
d8['theme_month_percentage'] = d8['theme_month']/d8['month_count']
```
